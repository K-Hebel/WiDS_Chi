## The target is diabetes_mellitus = 1 
### I'd like to review all the data where the target is equal to 1

```python
df_yes=df[df.diabetes_mellitus == 1]
df_yes.head()
```

```python
#dataframe of patient vital statistics for diabetic patients


#segregate vitals columns from the data_dict
vitals_cols= data_dict[data_dict['Category'].isin(['vitals', 'gas', 'labs blood gas'])]
vitals_cols=list(vitals_cols['Variable Name'].values)
vitals_cols

#create dataframe of only vitals columns

df_vitals = df_yes[vitals_cols]
df_vitals.describe()
```

```python
#segregate APACHE covariate and APACHE comorbidity columns from the data_dict
APACHE_cols= data_dict[data_dict['Category'].isin(['APACHE covariate', 'APACHE comorbidity'])]
APACHE_cols=list(APACHE_cols['Variable Name'].values)
APACHE_cols

#create dataframe of only APACHE columns

df_APACHE = df_yes[APACHE_cols]
df_APACHE.describe()
```

```python
df_APACHE.head()
```

```python
#segregate Demographic covariate columns from the data_dict
demo_cols= data_dict[data_dict['Variable Name'].isin(['gender','height', 'ethnicity', 'age', 'weight'])]
demo_cols=list(demo_cols['Variable Name'].values)
demo_cols

#create dataframe of only APACHE columns

df_demo = df_yes[demo_cols]
df_demo.describe()
```
